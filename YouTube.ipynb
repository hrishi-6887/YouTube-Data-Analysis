```python
category_subscribers_mean=df.groupby("category")["subscribers"].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.plot(kind="bar")
plt.title("Mean Subscribe By Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")

#1e7 = 10 million
 
```

```python
channel_type_counts=df['channel_type'].value_counts()
channel_type_counts
```

```python
channel_type_counts.head(8).plot(kind='pie',autopct="%1.1f%%")
plt.title('Channel Type Distribution')
```

```python
country_subscribers_total=df.groupby("Country")["subscribers"].sum().sort_values(ascending=False)
```

```python
country_subscribers_total
```

```python
country_subscribers_total.head(5).plot(kind="bar")
plt.title("Total Subscribers by Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation=45)
```

```python
df.groupby("channel_type")["video_views_rank"].median().sort_values(ascending=False)
```

```python
cat_view=df.groupby("category")["video views"].mean().sort_values(ascending=0)
cat_view.head(10)
```

```python
df.groupby("channel_type")["uploads"].sum().sort_values(ascending = 0)
```

```python
# Distribution of earnings based on channel type

sns.boxplot(x="channel_type",y="highest_monthly_earnings",data=df,hue="channel_type", palette="bright",legend=False)
plt.title("Distribution of Highest monthly earnings by channel Type")
plt.xticks(rotation = 90)
plt.show()

 
```

```python
category_highest_monthly_earnings_sum=df.groupby("category")["highest_monthly_earnings"].sum()
category_highest_monthly_earnings_sum
```

```python
category_highest_monthly_earnings_sum.head().plot(kind="pie",autopct="%1.1f%%")
plt.title("Percentange of Highest Monthly Earnings By Category")
plt.ylabel("")
plt.show()
```

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

monthly_avg = df.groupby("created_month")["subscribers_for_last_30_days"].mean().reindex(month_order)
monthly_avg.plot(kind='line', marker="p")
plt.title("Average Subscribers for Last 30 Days by Month")
plt.ylabel("Avg Subscribers for Last 30 Days")
plt.xticks(ticks=range(len(monthly_avg)), labels=monthly_avg.index, rotation=45)
plt.show()
```
